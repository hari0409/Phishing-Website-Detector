```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv("phishing.csv")
```

```python
data.head(10)
```

```python
data.drop(["Index"],inplace=True,axis="columns")
data.head(10)
```

```python
df=data.drop(["class"],axis="columns")
target=data["class"]
df.columns
print(len(df.columns))
```

```python
df.head(10)
```

```python
target.head(10)
```

```python
target.reset_index(drop=True,inplace=True)
target.head(10)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(df,target,test_size=0.3)
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier(max_depth=50,n_estimators=50)
model.fit(X_train.values,y_train.values)
```

```python
model.score(X_test.values,y_test.values)
```

```python
index=89
print(model.predict([X_test.iloc[index].values]))
print(y_test.iloc[index])
```

```python
import joblib
filename="phishingDetect.sav"
joblib.dump(model,filename)
```
